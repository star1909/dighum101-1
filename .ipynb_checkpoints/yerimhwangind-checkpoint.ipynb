### > Lemmatization

`lemmatizer_function` takes in a list of tokens and lemmatizes each item.

```python
def lemmatizer_function(tokens):
    words = ' '.join(tokens)
    doc = nlp(words)
    return [token.lemma_ for token in doc]
```

```python
foxnews_lemm = lemmatizer_function(foxnews_nostops)
huffpost_lemm = lemmatizer_function(huffpost_nostops)
yahoo_lemm = lemmatizer_function(yahoo_nostops)
breitbart_lemm = lemmatizer_function(breitbart_nostops)
nytimes_lemm = lemmatizer_function(nytimes_nostops)
```

### > Removing extra punctuation

```python
def remove_punct2(tokens_list):
    lemm_tokens = tokens_list
    for token in lemm_tokens:
        if token in punctuation:
            lemm_tokens.remove(token)
        if token.isdigit():
            lemm_tokens.remove(token)
    return lemm_tokens
```

`word_frequency_len` returns the word count for the term specified.

```python
def word_frequency_len(df, word):
    return len(df[df["Vocab"] == word])
```

```python
barWidth = 0.38

trump = [word_frequency_len(huffpost_dataframe, "Trump"), word_frequency_len(nytimes_dataframe, "Trump"),\
         word_frequency_len(yahoo_dataframe, "Trump"), word_frequency_len(foxnews_dataframe, "Trump"),\
        word_frequency_len(breitbart_dataframe, "Trump")]
clinton = [word_frequency_len(huffpost_dataframe, "Clinton"), word_frequency_len(nytimes_dataframe, "Clinton"),\
         word_frequency_len(yahoo_dataframe, "Clinton"), word_frequency_len(foxnews_dataframe, "Clinton"),\
        word_frequency_len(breitbart_dataframe, "Clinton")]
# Set position of bar on X axis
r1 = np.arange(len(trump))
r2 = [x + barWidth for x in r1]
 
# Make the plot
plt.bar(r1, trump, color='red', width=barWidth, edgecolor='white', label='Trump')
plt.bar(r2, clinton, color='blue', width=barWidth, edgecolor='white', label='Clinton')
 
# Add xticks on the middle of the group bars
plt.xlabel('Media Source', fontweight='bold')
plt.ylabel('Word Count')
plt.xticks([r + barWidth for r in range(len(trump))], ['HuffPost', 'NY Times', 'Yahoo', 'Fox News', 'Breitbart'])
 
# Create legend & Show graphic
plt.legend()
plt.title('"Trump"/"Clinton" Word Count')
plt.box(on = None)
plt.show()

```

### Calculate sentiment for vocab and add to dataframe

`return_compoundscore` returns a list of compound sentiment scores for all values in the provided dataframe column

```python
def return_compoundscore(column):
    ee = []
    for i in column: 
        ee.append(i["compound"])
    return ee
```

Adding a `Sentiment` column that contains a dictionary of neg/neu/pos sentiment values for each individual term, and a `Compound Sentiment` column that only contains the compound score.

```python
#foxnews
foxnews_dataframe["Sentiment"] = foxnews_dataframe["Vocab"].apply(sid.polarity_scores)
foxnews_dataframe["Compound Sentiment"] = return_compoundscore(foxnews_dataframe["Sentiment"])

#huffpost
huffpost_dataframe["Sentiment"] = huffpost_dataframe["Vocab"].apply(sid.polarity_scores)
huffpost_dataframe["Compound Sentiment"] = return_compoundscore(huffpost_dataframe["Sentiment"])

#yahoo
yahoo_dataframe["Sentiment"] = yahoo_dataframe["Vocab"].apply(sid.polarity_scores)
yahoo_dataframe["Compound Sentiment"] = return_compoundscore(yahoo_dataframe["Sentiment"])

#nytimes
nytimes_dataframe["Sentiment"] = nytimes_dataframe["Vocab"].apply(sid.polarity_scores)
nytimes_dataframe["Compound Sentiment"] = return_compoundscore(nytimes_dataframe["Sentiment"])

#breitbart
breitbart_dataframe["Sentiment"] = breitbart_dataframe["Vocab"].apply(sid.polarity_scores)
breitbart_dataframe["Compound Sentiment"] = return_compoundscore(breitbart_dataframe["Sentiment"])
```

```python
def grey_out(val):
    """
    Takes a scalar and returns a string with
    the css property `'color: red'` for negative
    strings, black otherwise.
    """
    color = 'lightgrey' if val > 0.7 else 'black'
    return 'color: %s' % color
```

The following analysis will explore the sentiment for each text according to the calculated compound sentiment value. 

`neg_pos` returns the cumulative sum of the negative and positive compound values in a list. 

`avg_neg_pos` returns the average negative and positive compound values in a list.

```python
def neg_pos(all_list):
    neg = 0
    pos = 0
    for i in all_list:
        if i <= 0: 
            neg += i
        elif i >= 0: 
            pos += i
    return [neg, pos]
```

There appears to be more of a difference in the expression of negative sentiment than positive sentiment in these articles, yet the values for each article are relatively similar with no clear pattern. 

```python
def avg_neg_pos(all_list):
    neg = 0
    neg_count = 0
    pos = 0
    pos_count = 0
    for i in all_list:
        if i <= 0: 
            neg += i
            neg_count += 1
        elif i >= 0: 
            pos += i
            pos_count += 1
    if neg_count == 0:
        avgneg = 0
        avgpos = pos/pos_count
    if pos_count == 0:
        avgpos = 0
        avgneg = neg/neg_count
    else:
        avgneg = neg/neg_count
        avgpos = pos/pos_count
    return [avgneg, avgpos]
            
```

### > Analysis by POS
In order to analyse the sentiment by parts of speech (POS), the spread of descriptive POS tags - adjectives and adverbs - are assessed.

`pos_df` returns a dataframe that contains only the terms with parts of speech from the list of POS acronyms provided.

```python
def pos_df(df, pos_list):
    agg = df[df["POS"] == pos_list[0]]
    for i in pos_list[1:]:
        agg.append(df[df["POS"] == i])
    return agg
```

## TFIDF

Term Frequency–Inverse Document Frequency (TFIDF) can be used to analyze the words in all of the articles and measure its importance to the full article.

```python
corpus = [to_string(huffpost_dataframe_cut["Vocab"]), to_string(nytimes_dataframe_cut["Vocab"]), 
          to_string(yahoo_dataframe_cut["Vocab"]), to_string(foxnews_dataframe_cut["Vocab"]),
          to_string(breitbart_dataframe_cut["Vocab"])]
```

```python
vectorizer = CountVectorizer()
vectorizer.fit(corpus);

vector = vectorizer.transform(corpus)
```

```python
tfidfvec = TfidfVectorizer()
sparse_tfidf = tfidfvec.fit_transform(corpus)

tfidf = pd.DataFrame(sparse_tfidf.toarray(), columns=tfidfvec.get_feature_names(), index = ["HuffPost", "NY Times", "Yahoo", "Fox News", "Breitbart"])
tfidf.head()
```

```python
colormap = sns.light_palette("orange", as_cmap=True)
tfidf_vis = tfidf.style.background_gradient(cmap=colormap)
tfidf_vis
```

By adding color to highlight the different levels of frequency, it can be seen that certain words are more heavily concentrated in certain types of text. For example "businessman" is used more often in the conservative texts, most likely used as a description of Donald Trump.

# Conclusion

Textual analysis of news media texts reveals that subtle pattern do exist in how an event is framed through media. Direct comparison of frequently used keywords revealed that journalists shape the story of the event through the language used. Conservative-leaning texts focus more on the details of the event itself, whereas liberal-leaning texts focus more on the external, background details regarding the "how" of the event. 
Analyzing the sentiment of the overall text does not showcase an accurate depiction of the mood of the articles. However, analyzing the sentiments according to parts of speech enabled visualizations that show the differences in emotional charges of words used. Through these small alterations in language, conservative-leaning media and liberal-leaning media construct and narrate a single event within the framings of implicit biases. 

It must be noted that the results of this analysis are not perfect. The data is dependent on the natural language processing tools that take words out of linguistic context, and are collected using only one article per news source, which is a small dataset and thus subject to error. The scope of news outlets included are also not comprehensive of all American news media. This project sets up the methods to expand upon this project topic for future development by incorporating a wider dataset for both articles analyzed and sources chosen.
